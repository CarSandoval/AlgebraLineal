# 5.1 Definición de valores propios y vectores propios

## 5.1.1 Definición de valores y vectores propios

Dada una matriz cuadrada $A$ de tamaño $n \times n$, un **valor propio** (o **autovalor**, o **eigenvalor**) es un número $\lambda$ tal que existe un **vector no nulo** $v$ (llamado **vector propio**) con la propiedad:

$$
A v = \lambda v
$$

En otras palabras, aplicar $A$ a $v$ simplemente lo **estira o invierte** por un factor $\lambda$, pero no cambia su dirección.


## 5.1.2 Cálculo de valores propios en $ \mathbb{R}^n $


1) **Punto de partida (definición):**  
Un escalar $ \lambda $ es un **valor propio** de una matriz cuadrada $ A \in \mathbb{R}^{n \times n} $ si existe un vector **no nulo** $ v \in \mathbb{R}^n $ tal que

$$
A v = \lambda v .
$$

2) **Reacomodo a un sistema homogéneo:**  
Llevamos todos los términos al mismo lado:

$$
A v - \lambda v = 0 \quad \Longrightarrow \quad (A - \lambda I) v = 0 .
$$


3) **Condición para soluciones no triviales:**  
El sistema homogéneo $ (A - \lambda I) v = 0 $ tiene una **solución no trivial** (es decir, $ v \neq 0 $) **si y solo si** la matriz $ A - \lambda I $ **no es invertible**.

4) **Criterio de invertibilidad vía determinante:**  
Una matriz cuadrada es invertible **si y solo si** su determinante es **no cero**. Por lo tanto, para que $ (A - \lambda I) $ **no** sea invertible (y existan autovectores no nulos) debe cumplirse:

$$
\det(A - \lambda I) = 0 .
$$




### Ejemplo


Sea

$$
A =
\begin{pmatrix}
2 & 1 & 0 \\
0 & 2 & 1 \\
1 & 0 & 2
\end{pmatrix}.
$$

Notamos que **la suma de cada renglón es 3**, así que es razonable probar $\lambda = 3$ como candidato a valor propio
, asociado al vector $\mathbf{1}=(1,1,1)^T$ (¿porqué?). Verifiquémoslo formalmente.



1) Construir $ A - \lambda I $ con $ \lambda = 3 $

$$
A - 3I =
\begin{pmatrix}
2-3 & 1   & 0 \\
0   & 2-3 & 1 \\
1   & 0   & 2-3
\end{pmatrix}
=
\begin{pmatrix}
-1 & 1 & 0 \\
0  & -1 & 1 \\
1  & 0  & -1
\end{pmatrix}.
$$



2) Calcular $ \det(A - 3I) $ (expansión por cofactores en la primera fila)

Sea

$$
M =
\begin{pmatrix}
-1 & 1 & 0 \\
0  & -1 & 1 \\
1  & 0  & -1
\end{pmatrix}.
$$

Entonces

$$
\det(M)
= (-1)\,\det\!\begin{pmatrix}-1 & 1 \\ 0 & -1\end{pmatrix}
\;-\; (1)\,\det\!\begin{pmatrix}0 & 1 \\ 1 & -1\end{pmatrix}
\;+\; (0)\, \det\!\begin{pmatrix}0 & -1 \\ 1 & 0\end{pmatrix}
$$

Calculamos los determinantes $ 2\times 2 $:
- $$\det\!\begin{pmatrix}-1 & 1 \\ 0 & -1\end{pmatrix} = (-1)(-1) - (1)(0) = 1.$$
- $$\det\!\begin{pmatrix}0 & 1 \\ 1 & -1\end{pmatrix} = (0)(-1) - (1)(1) = -1.$$

Sustituyendo:

$$
\det(M) = (-1)\cdot 1 \;-\; (1)\cdot(-1) + 0 = -1 + 1 + 0 = 0.
$$

Como

$$
\det(A - 3I) = 0,
$$

concluimos que $ \lambda = 3 $ es **valor propio** de $ A $.


3) (Opcional) Hallar un vector propio asociado

Resolvemos $ (A - 3I)v = 0 $ con

$$
\begin{pmatrix}
-1 & 1 & 0 \\
0  & -1 & 1 \\
1  & 0  & -1
\end{pmatrix}
\begin{pmatrix} x \\ y \\ z \end{pmatrix}
=
\begin{pmatrix} 0 \\ 0 \\ 0 \end{pmatrix}.
$$

De la primera ecuación: $ -x + y = 0 \Rightarrow y = x $.  
De la segunda: $ -y + z = 0 \Rightarrow z = y = x $.  
La tercera queda satisfecha automáticamente. Por tanto, una familia de autovectores es

$$
v = \begin{pmatrix} x \\ x \\ x \end{pmatrix} = x \begin{pmatrix} 1 \\ 1 \\ 1 \end{pmatrix},\quad x \neq 0.
$$

Verificación directa con $ v = (1,1,1)^T $:

$$
A v =
\begin{pmatrix}
2 & 1 & 0 \\
0 & 2 & 1 \\
1 & 0 & 2
\end{pmatrix}
\begin{pmatrix} 1 \\ 1 \\ 1 \end{pmatrix}
=
\begin{pmatrix} 3 \\ 3 \\ 3 \end{pmatrix}
= 3 \begin{pmatrix} 1 \\ 1 \\ 1 \end{pmatrix} = 3v.
$$

**Conclusión:** $ \lambda = 3 $ es valor propio entero de $ A $, que tiene entradas enteras, y un vector propio asociado es $ (1,1,1)^T $.
