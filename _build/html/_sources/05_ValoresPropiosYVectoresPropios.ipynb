# 5 Valores propios y vectores propios

# Introducción a los valores propios

Los **valores propios** (también llamados *autovalores*) son una herramienta fundamental del álgebra lineal que permiten entender cómo una **transformación lineal** o una **matriz** modifica los vectores en un espacio.  
Intuitivamente, un valor propio describe **cuánto se estira o se encoge** un vector especial (llamado *vector propio*) cuando se aplica una transformación.

Si pensamos en una matriz como una máquina que transforma vectores, los **vectores propios** son aquellos que **no cambian de dirección**, solo de tamaño, y el **valor propio** indica **cuánto se escalan**.  
Por ejemplo, si una matriz representa una rotación o una proyección, los valores propios nos indican si hay direcciones que permanecen fijas o que se amplifican.



## Aplicaciones en computación

En **ciencia de datos** y **computación**, los valores propios aparecen constantemente:
- En **reducción de dimensiones** (como en el método **PCA**, *Análisis de Componentes Principales*), los autovalores indican la **importancia o varianza** de cada componente.
- En **redes neuronales**, ayudan a analizar la **estabilidad** de los gradientes y la **propagación del error**.
- En **visión por computadora**, los valores propios permiten identificar **formas dominantes**, **direcciones principales** o **movimientos predominantes**.
- En **simulación física o ingeniería**, los autovalores indican **modos de vibración** o **frecuencias naturales** de un sistema.



## Aplicaciones en biomedicina

En **biomedicina**, los valores propios se usan para analizar sistemas complejos:
- En **electroencefalografía (EEG)** o **fMRI**, se emplean para extraer **patrones dominantes** de actividad cerebral.  
- En **análisis genómico**, ayudan a identificar **variaciones principales** entre muestras biológicas (por ejemplo, tipos de células o expresión de genes).
- En **modelado de sistemas fisiológicos**, permiten estudiar la **estabilidad dinámica** de sistemas como el corazón o la respiración.



## Predicción del futuro y matrices de Markov

En el contexto de **cadenas de Markov**, las matrices describen sistemas donde el **estado futuro** depende del **estado actual**.  
Si la matriz de transición no tiene valores propios nulos (es decir, es invertible), sus valores propios determinan **cómo evoluciona el sistema con el tiempo**:

- El **mayor valor propio (generalmente 1)** indica el **estado estable** o de equilibrio al que el sistema tiende.
- Los **otros valores propios** (menores en magnitud) indican **cómo de rápido** las probabilidades se aproximan a ese equilibrio.
- Por eso, los valores propios permiten **predecir el comportamiento futuro** del sistema sin simularlo paso a paso.


En resumen, los valores propios son una de las ideas más potentes del álgebra lineal:  
nos revelan las **direcciones fundamentales** y los **factores de cambio** de cualquier transformación lineal, con aplicaciones que van desde el análisis de datos hasta la modelación de fenómenos biológicos y la predicción de sistemas dinámicos.
