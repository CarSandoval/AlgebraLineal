# 3.7 Cambio de base

**Base Canonica**

En $\mathbb{R}^n$, la **base canónica** es el conjunto formado por los vectores estándar:

$$
\mathcal{B}_{\text{can}} = \{ e_1, e_2, \dots, e_n \}
$$

donde $e_i$ es el vector que tiene un $1$ en la posición $i$ y ceros en las demás.  
Por ejemplo, en $\mathbb{R}^3$:

$$
e_1 = (1,0,0), \quad e_2 = (0,1,0), \quad e_3 = (0,0,1)
$$

La base canónica tiene la propiedad de que las coordenadas de un vector respecto a ella son sus componentes usuales.

**Cambio de Base**

Un **cambio de base** es el proceso mediante el cual representamos un mismo vector o transformación lineal en diferentes bases de un espacio vectorial.  
El objetivo principal es traducir las coordenadas de un vector desde una base original $\mathcal{B}$ a una base nueva $\mathcal{B}'$.  
Esto es fundamental en álgebra lineal porque algunas bases hacen que ciertos problemas sean más fáciles de resolver o interpretar.


# Notación de cambio de base

- Si $[x]_{\mathcal{B}}$ son las coordenadas del vector $x$ en la base $\mathcal{B}$, y $[x]_{\mathcal{B}'}$ son sus coordenadas en la base $\mathcal{B}'$, entonces el **cambio de base** de $\mathcal{B}$ a $\mathcal{B}'$ se representa usando una **matriz de cambio de base** $P_{\mathcal{B}' \leftarrow \mathcal{B}}$.

- La relación es:

  $$
  [x]_{\mathcal{B}'} = P_{\mathcal{B}' \leftarrow \mathcal{B}} [x]_{\mathcal{B}}
  $$

  donde $P_{\mathcal{B}' \leftarrow \mathcal{B}}$ es la matriz que lleva coordenadas en la base $\mathcal{B}$ a coordenadas en la base $\mathcal{B}'$.

- **Si quieres expresar el cambio de coordenadas de una base cualquiera $\mathcal{B}$ a la base canónica**, la matriz de cambio de base se denota como:

  $$
  P_{\mathcal{E} \leftarrow \mathcal{B}}
  $$

  Esto significa: "La matriz que convierte las coordenadas desde la base $\mathcal{B}$ a la base canónica $\mathcal{E}$".

- **Si quieres expresar el cambio desde la base canónica $\mathcal{E}$ a otra base $\mathcal{B}$:**

  $$
  P_{\mathcal{B} \leftarrow \mathcal{E}}
  $$

  Esto significa: "La matriz que convierte las coordenadas desde la base canónica $\mathcal{E}$ a la base $\mathcal{B}$".



## 3.7.1 Definición de coordenadas dada una base $\mathcal{B}$

Dado un espacio vectorial $V$ sobre un campo $K$ y una base $\mathcal{B} = \{v_1, v_2, \ldots, v_n\}$ de $V$,  
**las coordenadas** de un vector $x \in V$ respecto a la base $\mathcal{B}$ son los escalares únicos $a_1, a_2, \ldots, a_n \in K$ tales que:

$$
x = a_1 v_1 + a_2 v_2 + \cdots + a_n v_n
$$

El vector de coordenadas de $x$ respecto a $\mathcal{B}$ se denota como:

$$
[x]_{\mathcal{B}} = 
\begin{pmatrix}
a_1 \\
a_2 \\
\vdots \\
a_n
\end{pmatrix}
$$



### Explicación.

- Para **cada** vector $x$ del espacio $V$, existe **una única** combinación lineal de los vectores de la base $\mathcal{B}$ que da como resultado $x$.
- Los coeficientes de esa combinación son **las coordenadas** de $x$ respecto a la base $\mathcal{B}$.
- El proceso de obtener estas coordenadas se llama "expresar $x$ en la base $\mathcal{B}$".


### Unicidad de la representación en coordenadas respecto a una base

**Enunciado:**  
Si $\mathcal{B} = \{v_1, v_2, ..., v_n\}$ es una base de un espacio vectorial $V$ y $x \in V$,  
entonces **existe una única** manera de expresar $x$ como combinación lineal de los vectores de $\mathcal{B}$:

$$
x = a_1 v_1 + a_2 v_2 + \cdots + a_n v_n
$$
con $a_1, ..., a_n$ únicos.



**Demostración:**

Supón que existen dos formas de escribir $x$ como combinación lineal de los vectores de la base:

$$
x = a_1 v_1 + a_2 v_2 + \cdots + a_n v_n \\
x = b_1 v_1 + b_2 v_2 + \cdots + b_n v_n
$$

Restando ambas igualdades:

$$
0 = (a_1 - b_1)v_1 + (a_2 - b_2)v_2 + \cdots + (a_n - b_n)v_n
$$

Pero los vectores $v_1, ..., v_n$ son **linealmente independientes** (por ser base),  
así que la única combinación lineal que da el vector cero es aquella donde **todos los coeficientes son cero**:

$$
a_1 - b_1 = 0 \\
a_2 - b_2 = 0 \\
\vdots \\
a_n - b_n = 0
$$

Por lo tanto, $a_i = b_i$ para todo $i=1, ..., n$.



**Conclusión:**  
Las coordenadas de cualquier vector respecto a una base están **bien definidas y son únicas**.



### Ejemplo:

Sea $\mathcal{B} = \{(1,1),\, (2,1)\}$ una base de $\mathbb{R}^2$ y $x = (5, 3)$.  
Queremos $a_1, a_2$ tales que:

$$
(5,3) = a_1 (1,1) + a_2 (2,1)
$$

Resolviendo:

$$
\begin{cases}
a_1 + 2a_2 = 5 \\
a_1 + a_2 = 3
\end{cases}
$$

Obtenemos $a_2 = 2$, $a_1 = 1$.

Por lo tanto,
  
$$
[x]_{\mathcal{B}} = 
\begin{pmatrix}
1 \\
2
\end{pmatrix}
$$



## 3.7.2 De canónica a no canónica

### Algoritmo

**Dado:**  
- Una base no canónica $\mathcal{B}' = \{v_1, v_2, \ldots, v_n\}$  
- Un vector $x$ en coordenadas canónicas (respecto a la base canónica $\mathcal{E}$)


**Desarrollo:**

Sea $P_{\mathcal{E} \leftarrow \mathcal{B}} = [v_1\ v_2\ \ldots\ v_n]$ (cada $v_i$ como columna). Entonces se debe cumplir que:

$$
P_{\mathcal{E}' \leftarrow \mathcal{B}'} [x]_{\mathcal{B}'} = [x]_{\mathcal{E}}
$$

Luego si multiplicamos por la izquierda de ambos lados de la ecuación por $(P_{\mathcal{E} \leftarrow \mathcal{B}'})^{-1}$ obtenemos que:

$$
\begin{align*}
(P_{\mathcal{E}' \leftarrow \mathcal{B}'})^{-1} P_{\mathcal{E} \leftarrow \mathcal{B}'} [x]_{\mathcal{B}'} = (P_{\mathcal{E}' \leftarrow \mathcal{B}'})^{-1} [x]_{\mathcal{E}}\\
[x]_{\mathcal{B}'} = (P_{\mathcal{E}' \leftarrow \mathcal{B}'})^{-1} [x]_{\mathcal{E}}
\end{align*}
$$

De aqui concluimos que $(P_{\mathcal{E}' \leftarrow \mathcal{B}'})^{-1}$ debe ser la matriz de cambio de base de $ [x]_\mathcal{E} \rightarrow [x]_\mathcal{B}' $ , por lo que:

$$
 (P_{\mathcal{B}' \leftarrow \mathcal{E}'}) = (P_{\mathcal{E}' \leftarrow \mathcal{B}'})^{-1} 
$$


**Pasos:**  
1. Forma la matriz de cambio de base $P_{\mathcal{E}' \leftarrow \mathcal{B}} = [v_1\ v_2\ \ldots\ v_n]$ (cada $v_i$ como columna).
2. Invertimos la matriz y obtenemos $P_{\mathcal{B}' \leftarrow \mathcal{E}} = (P_{\mathcal{E}' \leftarrow \mathcal{B}'})^{-1} $
2. Multiplicamos $P_{\mathcal{B}' \leftarrow \mathcal{E}}\ [x]_\mathcal{E} = [x]_\mathcal{B} $
3. El vector $[x]_{\mathcal{B}'}$ son las coordenadas de $x$ en la base $\mathcal{B}'$.



### Ejemplo numérico

Sea $\mathcal{B}' = \{ (1,1), (2,1) \}$ y $x = (5, 3) = [x]_{\mathcal{E}}$.

1. La matriz de cambio de base es:

   $$
   P_{\mathcal{E} \leftarrow \mathcal{B}'} = 
   \begin{pmatrix}
   1 & 2 \\
   1 & 1
   \end{pmatrix}
   $$

2. Su inversa es:

   $$
   P_{\mathcal{B}' \leftarrow \mathcal{E}} =
   \begin{pmatrix}
   1 & 2 \\
   1 & 1
   \end{pmatrix}^{-1}
   =
   \begin{pmatrix}
   -1 & 2 \\
   1 & -1
   \end{pmatrix}
   $$

3. Multiplicamos:

   $$
   [x]_{\mathcal{B}'} = P_{\mathcal{B}' \leftarrow \mathcal{E}} \ [x]_{\mathcal{E}}
   =
   \begin{pmatrix}
   -1 & 2 \\
   1 & -1
   \end{pmatrix}
   \begin{pmatrix}
   5 \\
   3
   \end{pmatrix}
   $$

   $$
   = \begin{pmatrix}
   -1 \cdot 5 + 2 \cdot 3 \\
   1 \cdot 5 - 1 \cdot 3
   \end{pmatrix}
   = \begin{pmatrix}
   -5 + 6 \\
   5 - 3
   \end{pmatrix}
   = \begin{pmatrix}
   1 \\
   2
   \end{pmatrix}
   $$

Por lo tanto,

$$
[x]_{\mathcal{B}'} = (1, 2)
$$




### Ejemplo con python usando matplotlib



```python
import numpy as np
import matplotlib.pyplot as plt

# Paso 1: Definir la base no canónica
v1 = np.array([1, 1])
v2 = np.array([2, 1])

# Paso 2: Matriz de cambio de base de la base no canónica a la canónica
P_E_to_Bp = np.column_stack([v1, v2])
print("P_{E <- B'} =\n", P_E_to_Bp)

# Paso 3: Calcular la inversa (de la canónica a la no canónica)
P_Bp_to_E = np.linalg.inv(P_E_to_Bp)
print("P_{B' <- E} =\n", P_Bp_to_E)

# Paso 4: Vector en la base canónica
x_E = np.array([5, 3])

# Paso 5: Cambiar coordenadas
x_Bp = P_Bp_to_E @ x_E
print("[x]_{B'} =", x_Bp)

# Dibujo
origin = np.zeros(2)
plt.figure(figsize=(6,6))

# Base no canónica
plt.quiver(*origin, *v1, angles='xy', scale_units='xy', scale=1, color='r', label='$v_1$')
plt.quiver(*origin, *v2, angles='xy', scale_units='xy', scale=1, color='b', label='$v_2$')
# Vector en base canónica
plt.quiver(*origin, *x_E, angles='xy', scale_units='xy', scale=1, color='g', label='$x$ en canónica')

# Coordenadas sobre la base no canónica (reconstrucción)
x_from_Bp = x_Bp[0]*v1 + x_Bp[1]*v2
plt.quiver(*origin, *x_from_Bp, angles='xy', scale_units='xy', scale=1, color='m', label='$x$ reconstruido')

plt.xlim(-1, 7)
plt.ylim(-1, 7)
plt.grid(True)
plt.legend()
plt.title("Cambio de base: canónica a base no canónica")
plt.show()
```

## 3.7.3 De no canónica a canónica

### Algoritmo

**Dado:**  
- Una base no canónica $\mathcal{B}' = \{v_1, v_2, \ldots, v_n\}$
- Coordenadas $[x]_{\mathcal{B}'} = (a_1, ..., a_n)$ respecto a $\mathcal{B}'$

**Pasos:**  
1. Forma la matriz de cambio de base $P_{\mathcal{E} \leftarrow \mathcal{B}'} = [v_1\ v_2\ \ldots\ v_n]$ (cada $v_i$ como columna).
2. Calcula el vector en la base canónica:

   $$
   [x]_{\mathcal{E}} = P_{\mathcal{E} \leftarrow \mathcal{B}'} \cdot [x]_{\mathcal{B}'}
   $$



### Ejemplo numérico

Sea $\mathcal{B}' = \{ (1,1), (2,1) \}$ y $[x]_{\mathcal{B}'} = (1, 2)$.

Entonces:

$$
[x]_{\mathcal{E}} = P_{\mathcal{E} \leftarrow \mathcal{B}'} \cdot [x]_{\mathcal{B}'} = 1 \cdot (1,1) + 2 \cdot (2,1) = (1+4,\ 1+2) = (5, 3)
$$



### Ejemplo con python usando matplotlib


```python
import numpy as np
import matplotlib.pyplot as plt

P = np.array([[1, 2],
              [1, 1]])
a = np.array([1, 2])

x = P @ a
print("Vector en la base canónica:", x)

origin = np.zeros(2)
v1 = P[:, 0]
v2 = P[:, 1]

plt.figure(figsize=(6,6))
plt.quiver(*origin, *v1, angles='xy', scale_units='xy', scale=1, color='r', label='$v_1$')
plt.quiver(*origin, *v2, angles='xy', scale_units='xy', scale=1, color='b', label='$v_2$')
plt.quiver(*origin, *x, angles='xy', scale_units='xy', scale=1, color='g', label='$x$')
plt.xlim(-1, 6)
plt.ylim(-1, 6)
plt.grid(True)
plt.legend()
plt.title('Cambio de base: no canónica a canónica')
plt.show()
```

## 3.7.4 De no canónica a no canónica


### Algoritmo

**Dado:**  
- Una base original $\mathcal{B} = \{u_1, ..., u_n\}$ y una base destino $\mathcal{B}' = \{v_1, ..., v_n\}$  
- Coordenadas $a$ respecto a $\mathcal{B}$

**Pasos:**  
1. Forma $Q = [u_1\ \ldots\ u_n]$ y $P = [v_1\ \ldots\ v_n]$  
2. Calcula $x = Q a$ (vector en base canónica)  
3. Resuelve $P b = x$ para $b$ (coordenadas en la base destino)



### Ejemplo numérico

Sean $\mathcal{B} = \{ (1,0), (0,2) \}$, $\mathcal{B}' = \{ (1,1), (2,1) \}$, $a = (3,2)$.

1. $x = Q a = 3 \cdot (1,0) + 2 \cdot (0,2) = (3, 4)$

2. Busca $b_1, b_2$ tales que:

$$
(3,4) = b_1(1,1) + b_2(2,1) = (b_1 + 2b_2,\, b_1 + b_2)
$$

Sistema:

$$
\begin{cases}
b_1 + 2b_2 = 3 \\
b_1 + b_2 = 4
\end{cases}
$$
Restando: $b_2 = -1$, $b_1 = 5$.



### Ejemplo con python usando matplotlib


```python
import numpy as np
import matplotlib.pyplot as plt

Q = np.array([[1, 0],
              [0, 2]])
P = np.array([[1, 2],
              [1, 1]])
a = np.array([3, 2])

x = Q @ a
b = np.linalg.solve(P, x)
print("Coordenadas en la base destino:", b)

origin = np.zeros(2)
u1 = Q[:, 0]
u2 = Q[:, 1]
v1 = P[:, 0]
v2 = P[:, 1]

plt.figure(figsize=(7,7))
plt.quiver(*origin, *u1, angles='xy', scale_units='xy', scale=1, color='purple', label='u1 (Base original)')
plt.quiver(*origin, *u2, angles='xy', scale_units='xy', scale=1, color='pink', label='u2 (Base original)')
plt.quiver(*origin, *v1, angles='xy', scale_units='xy', scale=1, color='r', label='v1 (Base destino)')
plt.quiver(*origin, *v2, angles='xy', scale_units='xy', scale=1, color='b', label='v2 (Base destino)')
plt.quiver(*origin, *x, angles='xy', scale_units='xy', scale=1, color='g', label='x (vector)')
plt.xlim(-1, 7)
plt.ylim(-1, 7)
plt.grid(True)
plt.legend()
plt.title('Cambio de base: no canónica a no canónica')
plt.show()
```
