# 5.3 Diagonalización de matrices


El concepto de diagonalización surge en el siglo XIX en el contexto del desarrollo del álgebra lineal y la teoría de ecuaciones diferenciales. Matemáticos como **Augustin-Louis Cauchy**, **Camille Jordan** y **Karl Weierstrass** estudiaron la estructura interna de las transformaciones lineales y la posibilidad de simplificarlas mediante cambios de base.

La idea central era encontrar una representación de una transformación lineal que facilitara los cálculos, especialmente al resolver sistemas de ecuaciones diferenciales lineales y al estudiar potencias de matrices. Este enfoque condujo naturalmente al estudio de los **valores propios**, **vectores propios** y, posteriormente, a la noción de diagonalización.

## 5.3.1 Definición de diagonalización

Sea $A \in \mathbb{R}^{n \times n}$ una matriz cuadrada. Se dice que $A$ es **diagonalizable** si existe una matriz invertible $P$ y una matriz diagonal $D$ tales que

$$
A = P D P^{-1}.
$$

La matriz $D$ contiene en su diagonal los valores propios de $A$, mientras que las columnas de $P$ son vectores propios linealmente independientes asociados a dichos valores propios.

Equivalentemente, una matriz $A$ es diagonalizable si y solo si **existe una base del espacio formada por vectores propios de $A$**.


## 5.3.2 Justificación

Recordemos que un vector no nulo $\mathbf{v}$ es un **vector propio** de la matriz $A$ asociado al valor propio $\lambda$ si

$$
A\mathbf{v} = \lambda \mathbf{v}.
$$

Esta ecuación nos dice que la acción de la matriz $A$ sobre el vector $\mathbf{v}$ consiste únicamente en **escalarlo** por el factor $\lambda$.


### Construcción de la matriz $P$

Supongamos que la matriz $A \in \mathbb{R}^{n \times n}$ tiene $n$ vectores propios linealmente independientes:

$$
\mathbf{v}_1, \mathbf{v}_2, \ldots, \mathbf{v}_n,
$$

con valores propios asociados

$$
\lambda_1, \lambda_2, \ldots, \lambda_n.
$$

Construimos la matriz $P$ colocando estos vectores propios como columnas:

$$
P =
\begin{pmatrix}
| & | & & | \\
\mathbf{v}_1 & \mathbf{v}_2 & \cdots & \mathbf{v}_n \\
| & | & & |
\end{pmatrix}.
$$

Debido a que los vectores propios son linealmente independientes, la matriz $P$ es invertible.



### Construcción de la matriz diagonal $D$

Definimos la matriz diagonal $D$ colocando los valores propios correspondientes en la diagonal principal:

$$
D =
\begin{pmatrix}
\lambda_1 & 0 & \cdots & 0 \\
0 & \lambda_2 & \cdots & 0 \\
\vdots & \vdots & \ddots & \vdots \\
0 & 0 & \cdots & \lambda_n
\end{pmatrix}.
$$



### Relación fundamental: $AP = PD$

Consideremos el producto $AP$. Usando la definición de $P$, tenemos:

$$
AP =
\begin{pmatrix}
A\mathbf{v}_1 & A\mathbf{v}_2 & \cdots & A\mathbf{v}_n
\end{pmatrix}.
$$

Como cada $\mathbf{v}_i$ es un vector propio,

$$
A\mathbf{v}_i = \lambda_i \mathbf{v}_i.
$$

Por lo tanto,

$$
AP =
\begin{pmatrix}
\lambda_1 \mathbf{v}_1 & \lambda_2 \mathbf{v}_2 & \cdots & \lambda_n \mathbf{v}_n
\end{pmatrix}.
$$

Este producto puede escribirse de forma matricial como

$$
AP = P D.
$$

Esta igualdad es el **paso clave** de toda la diagonalización.


### Despeje de la matriz $A$

A partir de la relación

$$
AP = P D,
$$

multiplicamos ambos lados por $P^{-1}$ por la derecha:

$$
AP P^{-1} = P D P^{-1}.
$$

Como $P P^{-1} = I$, se obtiene finalmente

$$
A = P D P^{-1}.
$$



### Interpretación conceptual

La expresión

$$
A = P D P^{-1}
$$

puede interpretarse de la siguiente manera:

- $P^{-1}$ cambia las coordenadas al sistema definido por los vectores propios.
- $D$ aplica un escalamiento simple en cada dirección propia.
- $P$ regresa al sistema de coordenadas original.

Es decir, **la matriz $A$ actúa como una transformación complicada en la base estándar, pero como una transformación simple (diagonal) en la base de vectores propios**.



### Conclusión

La diagonalización es posible precisamente porque los vectores propios forman una base del espacio. En ese caso, la matriz $A$ puede descomponerse como

$$
A = P D P^{-1},
$$

lo que simplifica notablemente su análisis algebraico, geométrico y computacional.

## 5.3.3 Ejemplo númerico

Consideremos la matriz

$$
A =
\begin{pmatrix}
2 & 1 \\
1 & 2
\end{pmatrix}.
$$

Del análisis previo sabemos que sus valores propios son

$$
\lambda_1 = 3, \qquad \lambda_2 = 1,
$$

con vectores propios asociados

$$
\mathbf{v}_1 =
\begin{pmatrix}
1 \\
1
\end{pmatrix},
\qquad
\mathbf{v}_2 =
\begin{pmatrix}
1 \\
-1
\end{pmatrix}.
$$

Estos vectores son linealmente independientes, por lo que forman una base de $\mathbb{R}^2$. Definimos entonces

$$
P =
\begin{pmatrix}
1 & 1 \\
1 & -1
\end{pmatrix},
\qquad
D =
\begin{pmatrix}
3 & 0 \\
0 & 1
\end{pmatrix}.
$$

Se verifica que

$$
A = P D P^{-1},
$$

por lo que la matriz $A$ es diagonalizable.

## 5.3.4 Algunas consideraciones

Hasta ahora hemos visto que una matriz $A$ es diagonalizable cuando existe una base del espacio formada por **vectores propios**. Sin embargo, esta condición no siempre se cumple. En esta sección analizamos qué ocurre cuando los vectores propios **no son suficientes** o cuando un valor propio aparece con **multiplicidad mayor que uno**.



### Multiplicidad algebraica

Sea $A \in \mathbb{R}^{n \times n}$ y sea $\lambda$ un valor propio de $A$. La **multiplicidad algebraica** de $\lambda$ se define como el número de veces que $\lambda$ aparece como raíz del polinomio característico

$$
p_A(\lambda) = \det(A - \lambda I).
$$

Por ejemplo, si

$$
p_A(\lambda) = (\lambda - 2)^2(\lambda - 1),
$$

entonces el valor propio $\lambda = 2$ tiene multiplicidad algebraica $2$, mientras que $\lambda = 1$ tiene multiplicidad algebraica $1$.


### Multiplicidad geométrica

La **multiplicidad geométrica** de un valor propio $\lambda$ se define como la dimensión del subespacio de vectores propios asociado a $\lambda$, es decir,

$$
\text{dim}(\ker(A - \lambda I)).
$$

Este subespacio se denomina **espacio propio** asociado a $\lambda$.


### Relación entre ambas multiplicidades

Para cualquier matriz cuadrada $A$ y cualquier valor propio $\lambda$, siempre se cumple que

$$
1 \leq \text{multiplicidad geométrica}(\lambda)
\leq \text{multiplicidad algebraica}(\lambda).
$$

Sin embargo, ambas multiplicidades no necesariamente coinciden. **¿Por qué?**

### Multiplicidad geométrica menor a la multiplicidad algebraica

Una matriz $A$ **no es diagonalizable** cuando **no tiene suficientes vectores propios linealmente independientes** para formar una base del espacio.

Equivalentemente, $A$ no es diagonalizable si existe al menos un valor propio $\lambda$ tal que

$$
\text{multiplicidad geométrica}(\lambda)
< \text{multiplicidad algebraica}(\lambda).
$$

En este caso, no es posible construir una matriz $P$ invertible cuyas columnas sean vectores propios, y por lo tanto **no existe una descomposición** de la forma

$$
A = P D P^{-1}.
$$


### Ejemplo de matriz no diagonalizable

Consideremos la matriz

$$
A =
\begin{pmatrix}
1 & 1 \\
0 & 1
\end{pmatrix}.
$$

Su polinomio característico es

$$
p_A(\lambda) = (1 - \lambda)^2,
$$

por lo que el valor propio $\lambda = 1$ tiene multiplicidad algebraica $2$.

Ahora calculamos los vectores propios resolviendo

$$
(A - I)\mathbf{v} = \mathbf{0}.
$$

Se obtiene

$$
A - I =
\begin{pmatrix}
0 & 1 \\
0 & 0
\end{pmatrix}.
$$

El sistema asociado produce una única dirección independiente de vectores propios, por ejemplo

$$
\mathbf{v} =
\begin{pmatrix}
1 \\
0
\end{pmatrix}.
$$

Por lo tanto, la multiplicidad geométrica de $\lambda = 1$ es $1$, menor que su multiplicidad algebraica.

Esta matriz **no es diagonalizable**.


Cuando una matriz no es diagonalizable, la transformación lineal asociada **no puede descomponerse completamente en escalamiento independiente de direcciones**. En lugar de ello, aparecen efectos de *acoplamiento* entre direcciones, como deslizamientos o deformaciones adicionales.

Geométricamente, esto significa que **no existen suficientes direcciones invariantes** para describir la transformación de manera puramente diagonal.



### Ejemplos graficos en python

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Caso 1: NO diagonalizable (A NO diagonal)
# -----------------------------
A = np.array([[1, 1, 0],
              [0, 1, 1],
              [0, 0, 2]], dtype=float)

rng = np.random.default_rng(1)
X = rng.uniform(-1, 1, size=(400, 3))
Y = (A @ X.T).T

w, V = np.linalg.eig(A)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_title("Caso 2: NO diagonalizable (mg < ma para λ repetido)")

ax.scatter(X[:,0], X[:,1], X[:,2], s=8, alpha=0.35, label="Original")
ax.scatter(Y[:,0], Y[:,1], Y[:,2], s=8, alpha=0.35, label="Transformada A·x")

# Dibujamos direcciones propias numéricas (ojo: para el λ repetido, puede salir 1 sola dirección independiente)
# Para hacerlo más claro, dibujamos el eigenspace de λ=1 resolviendo (A-I)v=0 con SVD
I = np.eye(3)
M = A - 1*I
# vector en el núcleo de M:
_, _, vh = np.linalg.svd(M)
v1 = vh[-1, :]
v1 = v1 / np.linalg.norm(v1)
ax.quiver(0,0,0, v1[0], v1[1], v1[2], length=2.0, normalize=True)
ax.text(1.9*v1[0], 1.9*v1[1], 1.9*v1[2], "eigenspace λ=1 (solo 1 dir.)")

# También dibujamos un eigenvector para λ=2
# (tomamos uno de los que devuelve eig)
idx2 = np.argmin(np.abs(w - 2))
v2 = np.real(V[:, idx2])
v2 = v2 / np.linalg.norm(v2)
ax.quiver(0,0,0, v2[0], v2[1], v2[2], length=2.0, normalize=True)
ax.text(1.9*v2[0], 1.9*v2[1], 1.9*v2[2], "dir. propia λ=2")

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.legend()
plt.show()

print("A =\n", A)
print("Autovalores =", w)
```

### Multiplicidad geométrica igual a multiplicidad algebraica

Consideremos ahora un ejemplo en el que un valor propio tiene **multiplicidad algebraica mayor o igual a 2**, pero aun así la matriz **sí es diagonalizable**, debido a que la **multiplicidad geométrica coincide** con la algebraica.


### Matriz de estudio

Sea la matriz

$$
A =
\begin{pmatrix}
2 & 0 & 0 \\
0 & 2 & 0 \\
0 & 0 & 3
\end{pmatrix}.
$$



### Cálculo del polinomio característico

El polinomio característico se obtiene como

$$
p_A(\lambda) = \det(A - \lambda I).
$$

Como $A$ es diagonal, se tiene directamente

$$
p_A(\lambda)
= (2 - \lambda)^2 (3 - \lambda).
$$

Por lo tanto:
- el valor propio $\lambda = 2$ tiene **multiplicidad algebraica 2**,
- el valor propio $\lambda = 3$ tiene **multiplicidad algebraica 1**.



### Cálculo de la multiplicidad geométrica

#### Valor propio $\lambda = 2$

Calculamos el espacio propio resolviendo

$$
(A - 2I)\mathbf{v} = \mathbf{0}.
$$

Se obtiene

$$
A - 2I =
\begin{pmatrix}
0 & 0 & 0 \\
0 & 0 & 0 \\
0 & 0 & 1
\end{pmatrix}.
$$

El sistema asociado es

$$
\begin{cases}
z = 0.
\end{cases}
$$

Las variables $x$ y $y$ son libres, por lo que el espacio propio está generado por

$$
\mathbf{v}_1 =
\begin{pmatrix}
1 \\ 0 \\ 0
\end{pmatrix},
\qquad
\mathbf{v}_2 =
\begin{pmatrix}
0 \\ 1 \\ 0
\end{pmatrix}.
$$

Así, la **multiplicidad geométrica** de $\lambda = 2$ es

$$
\text{dim}(\ker(A - 2I)) = 2.
$$



#### Valor propio $\lambda = 3$

De manera análoga,

$$
A - 3I =
\begin{pmatrix}
-1 & 0 & 0 \\
0 & -1 & 0 \\
0 & 0 & 0
\end{pmatrix}.
$$

El sistema conduce a

$$
x = 0, \quad y = 0,
$$

por lo que el espacio propio está generado por

$$
\mathbf{v}_3 =
\begin{pmatrix}
0 \\ 0 \\ 1
\end{pmatrix}.
$$

La multiplicidad geométrica de $\lambda = 3$ es $1$.



### Conclusión

En este ejemplo se cumple que:

- Para $\lambda = 2$:
  $$
  \text{multiplicidad algebraica} = 2,
  \qquad
  \text{multiplicidad geométrica} = 2.
  $$
- Para $\lambda = 3$:
  $$
  \text{multiplicidad algebraica} = 1,
  \qquad
  \text{multiplicidad geométrica} = 1.
  $$

Como **para todos los valores propios ambas multiplicidades coinciden**, la matriz $A$ es **diagonalizable**.



### Interpretación conceptual

Este ejemplo muestra que un valor propio repetido **no implica** pérdida de diagonalización. Lo realmente importante es que el espacio propio tenga **suficiente dimensión** para proporcionar una base del espacio.

En este caso, los vectores propios forman la base canónica de $\mathbb{R}^3$, y la matriz ya se encuentra en forma diagonal.


### Verificación con Python


```python
import numpy as np

A = np.array([
    [2, 0, 0],
    [0, 2, 0],
    [0, 0, 3]
])

valores, vectores = np.linalg.eig(A)

valores, vectores
```

### Ejemplo gráfico en python

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Caso 2: diagonalizable (A NO diagonal)
# -----------------------------
P = np.array([[1, 1, 0],
              [0, 1, 1],
              [1, 0, 1]], dtype=float)

D = np.diag([2, 2, 3])  # λ=2 tiene mult. algebraica 2
A = P @ D @ np.linalg.inv(P)  # A no diagonal pero diagonalizable

# Puntos del cubo [-1,1]^3
rng = np.random.default_rng(0)
X = rng.uniform(-1, 1, size=(400, 3))
Y = (A @ X.T).T

# Autovalores y autovectores
w, V = np.linalg.eig(A)  # columnas de V son vectores propios

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_title("Caso 1: diagonalizable (mg = ma para λ repetido)")

# nube original y transformada
ax.scatter(X[:,0], X[:,1], X[:,2], s=8, alpha=0.35, label="Original")
ax.scatter(Y[:,0], Y[:,1], Y[:,2], s=8, alpha=0.35, label="Transformada A·x")

# dibuja direcciones propias (normalizadas)
for i in range(3):
    v = np.real(V[:, i])
    v = v / np.linalg.norm(v)
    ax.quiver(0,0,0, v[0], v[1], v[2], length=2.0, normalize=True)
    ax.text(1.9*v[0], 1.9*v[1], 1.9*v[2], f"v{i+1}")

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.legend()
plt.show()

print("A =\n", A)
print("Autovalores =", w)
```

### Resumen

- La multiplicidad algebraica cuenta cuántas veces aparece un valor propio en el polinomio característico.
- La multiplicidad geométrica mide cuántos vectores propios independientes existen.
- Una matriz es diagonalizable **si y solo si** ambas multiplicidades coinciden para todos los valores propios.
- Cuando no coinciden, la matriz no es diagonalizable y requiere herramientas más avanzadas para su análisis.
