# 3.8 Bases ortonormales

Una **base ortonormal** en un espacio vectorial es una base cuyos vectores son todos de **norma 1** y **mutuamente ortogonales** (el producto punto entre ellos es cero si son diferentes).  
Este tipo de bases simplifica cálculos, permite interpretar coordenadas como proyecciones y es fundamental en álgebra lineal, geometría, análisis de señales, compresión de datos y algoritmos de machine learning.


## 3.8.1 Producto punto

### Definición
Dados dos vectores $u = (u_1, ..., u_n)$ y $v = (v_1, ..., v_n)$ en $\mathbb{R}^n$,  
el **producto punto** se define como:
$$
u \cdot v = u_1 v_1 + u_2 v_2 + \cdots + u_n v_n
$$



### Significado geométrico del producto punto

Geométricamente, el producto punto mide **cuánto "apuntan" dos vectores en la misma dirección**.  
Si el producto punto es:
- **positivo**, los vectores forman un ángulo agudo;
- **cero**, son ortogonales (perpendiculares);
- **negativo**, forman un ángulo obtuso.

El producto punto también se relaciona con la proyección de un vector sobre otro.

### Código en python usando matplotlib que lo muestre



```python
import numpy as np
import matplotlib.pyplot as plt

u = np.array([2, 1])
v = np.array([1, 2])
dot = np.dot(u, v)

origin = np.zeros(2)
plt.figure(figsize=(6,6))
plt.quiver(*origin, *u, angles='xy', scale_units='xy', scale=1, color='r', label='$u$')
plt.quiver(*origin, *v, angles='xy', scale_units='xy', scale=1, color='b', label='$v$')
plt.grid(True)
plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.legend()
plt.title(f'Producto punto: $u \\cdot v$ = {dot}')
plt.show()
```

### Ejemplos 

Sea $u = (1,2,3)$ y $v = (4,5,6)$.
$$
u \cdot v = 1\cdot4 + 2\cdot5 + 3\cdot6 = 4 + 10 + 18 = 32
$$

Sea $a = (2,0,-1)$ y $b = (1,3,2)$.
$$
a \cdot b = 2\cdot1 + 0\cdot3 + (-1)\cdot2 = 2 + 0 - 2 = 0
$$

### Propiedades del producto punto

- $u \cdot v = v \cdot u$ (conmutativo)
- $u \cdot (v + w) = u \cdot v + u \cdot w$ (distributivo)
- $(\alpha u) \cdot v = \alpha (u \cdot v)$ (compatibilidad con escalares)
- $u \cdot u \geq 0$, y $u \cdot u = 0$ si y solo si $u = 0$ (positividad)

## 3.8.2 Norma de un vector

### Definición general de norma

Una **norma** en un espacio vectorial $V$ sobre $\mathbb{R}$ (o $\mathbb{C}$) es una función  
$\| \cdot \| : V \to \mathbb{R}$ que asigna a cada vector un número real no negativo, y satisface tres propiedades fundamentales para todo $u, v \in V$ y todo escalar $\alpha$:

1. **Positividad:**  
   $$
   \|v\| \geq 0, \quad \text{y} \quad \|v\| = 0 \iff v = 0
   $$

2. **Homogeneidad absoluta:**  
   $$
   \|\alpha v\| = |\alpha| \, \|v\|
   $$

3. **Desigualdad triangular:**  
   $$
   \|u + v\| \leq \|u\| + \|v\|
   $$

### Propiedades de una norma

Supón que $\|\cdot\|$ es una norma sobre un espacio vectorial $V$.  
Entonces, a partir de los axiomas, se pueden demostrar las siguientes propiedades:

#### 1. Desigualdad del triángulo invertido (reverse triangle inequality)
Para todo $u, v \in V$,
$$
|\|u\| - \|v\|| \leq \|u - v\|
$$

**Demostración:**  
Por la desigualdad triangular,
$$
\|u\| = \|(u - v) + v\| \leq \|u - v\| + \|v\| \implies \|u\| - \|v\| \leq \|u - v\|
$$
Intercambiando $u$ y $v$, también se obtiene
$$
\|v\| - \|u\| \leq \|v - u\| = \|u - v\|
$$
Por lo tanto,
$$
|\|u\| - \|v\|| \leq \|u - v\|
$$


#### 2. Subaditividad para sumas finitas
Para cualquier colección finita de vectores $v_1, ..., v_k$ en $V$,
$$
\left\|\sum_{i=1}^k v_i \right\| \leq \sum_{i=1}^k \|v_i\|
$$



#### 3. La norma del opuesto es igual
Para cualquier $v \in V$,
$$
\| -v \| = \| v \|
$$





### Definición de la norma en $\mathbb{R}^n$

La norma **euclidiana** en $\mathbb{R}^n$ se define, para $v = (v_1, ..., v_n)$, como:
$$
\|v\| = \sqrt{v_1^2 + v_2^2 + \cdots + v_n^2}
$$


### Ejemplo numérico

Para $v = (1,2,2)$ en $\mathbb{R}^3$,
$$
\|v\| = \sqrt{1^2 + 2^2 + 2^2} = \sqrt{1 + 4 + 4} = \sqrt{9} = 3
$$



### Demostración de que la norma en $\mathbb{R}^n$ es una norma

Verifiquemos las tres propiedades para la norma euclidiana:

1. **Positividad:**  
   $$\|v\| = \sqrt{v_1^2 + \cdots + v_n^2} \geq 0$$  
   La raíz cuadrada de una suma de cuadrados nunca es negativa, y es cero solo si todos los $v_i$ son cero.

2. **Homogeneidad absoluta:**  
   Para escalar $\alpha$,
   $$
   \|\alpha v\| = \sqrt{(\alpha v_1)^2 + \cdots + (\alpha v_n)^2} = \sqrt{\alpha^2(v_1^2 + \cdots + v_n^2)} = |\alpha| \sqrt{v_1^2 + \cdots + v_n^2} = |\alpha| \|v\|
   $$

3. **Desigualdad triangular:**  
   $$
   \|u + v\|^2 = \sum_{i=1}^n (u_i + v_i)^2 = \sum_{i=1}^n u_i^2 + 2u_iv_i + v_i^2
   $$
   Usando desigualdad de Cauchy-Schwarz:
   $$
   \|u+v\| \leq \|u\| + \|v\|
   $$
   



### Normalización de vectores

La **normalización** de un vector $v \neq 0$ es obtener el vector unitario (norma 1) en la misma dirección:

$$
u = \frac{v}{\|v\|}
$$



### Ejemplo numérico

Para $v = (3,4)$,
$$
\|v\| = 5 \\
u = \frac{(3,4)}{5} = \left(\frac{3}{5}, \frac{4}{5}\right)
$$
Verifica:
$$
\|u\| = \sqrt{(3/5)^2 + (4/5)^2} = \sqrt{9/25 + 16/25} = \sqrt{1} = 1
$$



### Ejemplo con python


```python
import numpy as np
import matplotlib.pyplot as plt

v = np.array([3, 4])
norm_v = np.linalg.norm(v)
u = v / norm_v

print("Norma de v:", norm_v)
print("Vector normalizado:", u)
print("Norma del vector normalizado:", np.linalg.norm(u))

plt.figure(figsize=(7,7))

# Vector original (azul)
plt.arrow(0, 0, v[0], v[1], head_width=0.2, head_length=0.3, fc='b', ec='b', label='v original', linewidth=2, length_includes_head=True)
# Vector normalizado (rojo, más grueso)
plt.arrow(0, 0, u[0], u[1], head_width=0.15, head_length=0.2, fc='r', ec='r', label='v normalizado (norma 1)', linewidth=2, length_includes_head=True)

# Círculo unitario
circle = plt.Circle((0, 0), 1, color='g', fill=False, linestyle='--', linewidth=2, label='Círculo unitario')
plt.gca().add_artist(circle)

# Etiquetas de las puntas
plt.text(v[0]+0.1, v[1]+0.1, 'v', color='b', fontsize=14)
plt.text(u[0]+0.1, u[1]+0.1, 'u', color='r', fontsize=14)

# Ajustar ejes para ver todo claramente
max_val = max(norm_v, 1) + 1
plt.xlim(-1, max(v[0], 1)+2)
plt.ylim(-1, max(v[1], 1)+2)
plt.axis('equal')
plt.grid(True)
plt.legend(['v original', 'v normalizado (norma 1)', 'Círculo unitario'])
plt.title(r"Normalización de un vector en $\mathbb{R}^2$")
plt.show()
```

## 3.8.2 Angulo entre vectores

### Construcción de la fórmula del coseno de un ángulo entre vectores usando proyecciones y el Teorema de Pitágoras

Sean $u, v \in \mathbb{R}^2$, ambos distintos de cero, y sea $\theta$ el ángulo entre ellos.  
Supongamos que colocamos el origen en el inicio de ambos vectores.

1. **Construcción geométrica:**
   - Considera el triángulo formado por los puntos $O$ (origen), $A$ (punta de $u$), y $B$ (punta de $v$).
   - El vector $u - v$ es el tercer lado de ese triángulo.

2. **Por la Ley del Coseno tenemos:**
   $$
   |u - v|^2 = |u|^2 + |v|^2 - 2|u||v|\cos\theta
   $$

3. **Pero $|u - v|^2$ es también el producto punto:**
   $$
   |u - v|^2 = (u - v) \cdot (u - v) = u \cdot u - 2u \cdot v + v \cdot v = |u|^2 - 2u \cdot v + |v|^2
   $$

4. **Igualando las dos expresiones para $|u - v|^2$:**
   $$
   |u|^2 - 2u \cdot v + |v|^2 = |u|^2 + |v|^2 - 2|u||v|\cos\theta
   $$

5. **Se simplifican $|u|^2$ y $|v|^2$ de ambos lados:**
   $$
   -2u \cdot v = -2|u||v|\cos\theta
   $$

6. **Dividiendo entre $-2$ (y suponiendo que $u,v \neq 0$):**
   $$
   u \cdot v = |u||v|\cos\theta
   $$
   que es la fórmula buscada.



Por lo tanto,
$$
\cos\theta = \frac{u \cdot v}{|u||v|}
$$


### Construcción en python que de apoyo visual a la construcción anterior usando matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

# Vectores
u = np.array([2, 1])
v = np.array([1, 2])

# Ángulos respecto al eje x
angle_u = np.arctan2(u[1], u[0])
angle_v = np.arctan2(v[1], v[0])

# Asegúrate de que angle_v > angle_u (para el arco)
if angle_v < angle_u:
    angle_v, angle_u = angle_u, angle_v

arc = np.linspace(angle_u, angle_v, 100)
arc_x = 0.6 * np.cos(arc)
arc_y = 0.6 * np.sin(arc)

plt.figure(figsize=(7,7))
plt.arrow(0, 0, u[0], u[1], head_width=0.1, head_length=0.15, fc='r', ec='r', label='u', linewidth=2, length_includes_head=True)
plt.arrow(0, 0, v[0], v[1], head_width=0.1, head_length=0.15, fc='b', ec='b', label='v', linewidth=2, length_includes_head=True)
plt.arrow(v[0], v[1], u[0]-v[0], u[1]-v[1], head_width=0.1, head_length=0.15, fc='g', ec='g', label='u-v', linewidth=2, length_includes_head=True)

# Etiquetas
plt.text(*(u+0.1), '$u$', color='r', fontsize=14)
plt.text(*(v+0.1), '$v$', color='b', fontsize=14)
plt.text(*(0.5*(u+v)+0.2), '$u-v$', color='g', fontsize=14)
plt.text(0, 0, '$O$', fontsize=14)

# Arco del ángulo real entre u y v
plt.plot(arc_x, arc_y, color='k')
# Etiqueta para theta (en el medio del arco)
theta_mid = (angle_u + angle_v)/2
plt.text(0.8*np.cos(theta_mid), 0.8*np.sin(theta_mid), r'$\theta$', fontsize=16)

plt.xlim(-0.5, 2.7)
plt.ylim(-0.5, 2.7)
plt.grid(True)
plt.axis('equal')
plt.legend(['u', 'v', 'u-v'])
plt.title("Demostración geométrica de $u \\cdot v = |u||v|\\cos\\theta$")
plt.show()
```

### Definición formal de coseno de un ángulo entre vectores en $\mathbb{R}^n$

Dados $u, v \in \mathbb{R}^n$, el coseno del ángulo $\theta$ entre ellos se define como:
$$
\cos\theta = \frac{u \cdot v}{\|u\| \, \|v\|}
$$
donde $\|u\| = \sqrt{u_1^2 + \cdots + u_n^2}$ es la norma euclidiana.



### Ejemplos numéricos

Sea $u = (1,0)$ y $v = (0,1)$, entonces:
$$
u \cdot v = 0 \\
\|u\| = 1,\quad \|v\| = 1 \\
\cos\theta = 0 \implies \theta = 90^\circ
$$

Sea $a = (1,2)$ y $b = (2,4)$, entonces:
$$
a \cdot b = 1\cdot2 + 2\cdot4 = 2 + 8 = 10 \\
\|a\| = \sqrt{1^2 + 2^2} = \sqrt{5} \\
\|b\| = \sqrt{2^2 + 4^2} = \sqrt{20} \\
\cos\theta = \frac{10}{\sqrt{5}\sqrt{20}} = 1 \implies \theta = 0^\circ
$$


### Aplicaciones del ángulo entre 2 vectores a diferentes disciplinas

- **Análisis de similitud:** El ángulo entre dos vectores en espacios de alta dimensión se usa como medida de similitud en minería de datos y aprendizaje automático ("cosine similarity").
- **Física:** En trabajo mecánico y proyección de fuerzas.
- **Estadística:** Para correlación entre variables normalizadas.
- **Procesamiento de imágenes:** Detección de orientación y patrones.


## 3.8.3 Ortogonalidad de vectores

### Definición de la ortogonalidad entre 2 vectores en $\mathbb{R}^n$

Dos vectores $u, v \in \mathbb{R}^n$ son **ortogonales** si su producto punto es cero:
$$
u \cdot v = 0
$$


### Ejemplos de vectores ortogonales

Sea $u = (1,2,1)$ y $v = (2,-1,0)$.
$$
u \cdot v = 1\cdot2 + 2\cdot(-1) + 1\cdot0 = 2 - 2 + 0 = 0
$$
Así, $u$ y $v$ son ortogonales.



### Los mismos ejemplos dibujados en matplotlib usando python

```python
import numpy as np
import matplotlib.pyplot as plt

u = np.array([1, 2])
v = np.array([-2, 1])

origin = np.zeros(2)
plt.figure(figsize=(6,6))
plt.quiver(*origin, *u, angles='xy', scale_units='xy', scale=1, color='r', label='$u$')
plt.quiver(*origin, *v, angles='xy', scale_units='xy', scale=1, color='b', label='$v$')
plt.grid(True)
plt.xlim(-3, 3)
plt.ylim(-1, 3)
plt.legend()
plt.title("Vectores ortogonales ($u \\cdot v = 0$)")
plt.show()
```

## 3.8.4 Proceso de ortonormalización de GramSchmidt


El proceso de Gram-Schmidt permite transformar un conjunto de vectores linealmente independientes en un conjunto ortonormal que genera el mismo subespacio vectorial.  
Esto es esencial para simplificar problemas en álgebra lineal, análisis numérico y muchas aplicaciones en ciencia e ingeniería.



### Proyección de un vector sobre otro

Dados $v$ y $u$ en $\mathbb{R}^n$ ($u \neq 0$), la **proyección** de $v$ sobre $u$ es:
$$
\text{proj}_u(v) = \frac{v \cdot u}{u \cdot u} \, u
$$

**¿Por qué es así?**  
Buscamos un escalar $\lambda$ tal que $w = \lambda u$ sea el "punto más cercano" a $v$ en la recta generada por $u$.  
La diferencia $v - \lambda u$ debe ser ortogonal a $u$:
$$
(v - \lambda u) \cdot u = 0 \implies v \cdot u - \lambda (u \cdot u) = 0 \implies \lambda = \frac{v \cdot u}{u \cdot u}
$$
Por lo tanto,
$$
\text{proj}_u(v) = \lambda u = \frac{v \cdot u}{u \cdot u} u
$$

---

### Ortogonalización de 2 vectores usando la proyección

Sean $v_1, v_2 \in \mathbb{R}^n$ linealmente independientes.

1. Sea $u_1 = v_1$.
2. Proyecta $v_2$ sobre $u_1$: $\text{proj}_{u_1}(v_2) = \frac{v_2 \cdot u_1}{u_1 \cdot u_1} u_1$.
3. Resta la proyección: $u_2 = v_2 - \text{proj}_{u_1}(v_2)$.

Ahora $u_1$ y $u_2$ son ortogonales.

---

### Ortogonalización de $n$ vectores (generalización)

Dado $v_1, ..., v_n$ linealmente independientes en $\mathbb{R}^n$:

1. Sea $u_1 = v_1$.
2. Para $k = 2, ..., n$:
   $$
   u_k = v_k - \sum_{j=1}^{k-1} \text{proj}_{u_j}(v_k)
   = v_k - \sum_{j=1}^{k-1} \frac{v_k \cdot u_j}{u_j \cdot u_j} u_j
   $$
3. Los vectores $u_1, ..., u_n$ así construidos son ortogonales.

---

### Algoritmo de Gram-Schmidt: ortogonalizar y normalizar

El **algoritmo completo** de Gram-Schmidt produce una base ortonormal:

- Para cada $k = 1, ..., n$:
  1. Calcula el vector ortogonal $u_k$ como arriba.
  2. Normaliza:  
     $$
     e_k = \frac{u_k}{\|u_k\|}
     $$
- El conjunto $\{e_1, ..., e_n\}$ es una **base ortonormal** del subespacio generado.

---

### Ejemplo en $\mathbb{R}^3$

Sea
$$
v_1 = (1,1,0),\quad v_2 = (1,0,1),\quad v_3 = (0,1,1)
$$

**Paso 1:**  
$$
u_1 = v_1 = (1,1,0)
$$

**Paso 2:**  
$$
\text{proj}_{u_1}(v_2) = \frac{v_2 \cdot u_1}{u_1 \cdot u_1} u_1 = \frac{1\cdot1 + 0\cdot1 + 1\cdot0}{1^2+1^2+0^2} (1,1,0) = \frac{1}{2}(1,1,0) = (0.5,0.5,0)
$$
$$
u_2 = v_2 - \text{proj}_{u_1}(v_2) = (1,0,1) - (0.5,0.5,0) = (0.5, -0.5, 1)
$$

**Paso 3:**  
$$
\text{proj}_{u_1}(v_3) = \frac{v_3 \cdot u_1}{u_1 \cdot u_1} u_1 = \frac{0\cdot1 + 1\cdot1 + 1\cdot0}{2}(1,1,0) = \frac{1}{2}(1,1,0) = (0.5,0.5,0)
$$
$$
\text{proj}_{u_2}(v_3) = \frac{v_3 \cdot u_2}{u_2 \cdot u_2} u_2
$$
Calculemos $v_3 \cdot u_2 = 0.5\cdot0 + 1\cdot(-0.5) + 1\cdot1 = 0 - 0.5 + 1 = 0.5$  
$u_2 \cdot u_2 = 0.5^2 + (-0.5)^2 + 1^2 = 0.25 + 0.25 + 1 = 1.5$

$$
\text{proj}_{u_2}(v_3) = \frac{0.5}{1.5}(0.5, -0.5, 1) = (\frac{1}{6}, -\frac{1}{6}, \frac{1}{3})
$$
$$
u_3 = v_3 - \text{proj}_{u_1}(v_3) - \text{proj}_{u_2}(v_3)
= (0,1,1) - (0.5,0.5,0) - (\frac{1}{6}, -\frac{1}{6}, \frac{1}{3})
= (-\frac{2}{3}, \frac{2}{3}, \frac{2}{3})
$$

**Ortonormalizando:**

- $e_1 = u_1 / \|u_1\| = (1/\sqrt{2}, 1/\sqrt{2}, 0)$
- $e_2 = u_2 / \|u_2\| = (1/\sqrt{6}, -1/\sqrt{6}, 2/\sqrt{6})$
- $e_3 = u_3 / \|u_3\| = (-1/\sqrt{3}, 1/\sqrt{3}, 1/\sqrt{3})$

---

### Apoyo visual usando matplotlib
