# 1.3 Existencia y unicidad de la solución de ecuaciones lineales

Existen muchas preguntas a resolver cuando trabajamos con sisstemas de ecuaciones lineales. Una de las preguntas a las que frecuentemente nos enfrentaremos es: 
Dado que tenemos un sistema de ecuaciones lineales. ¿Cómo sabemos si existe alguna solucion? más aún, si existe una solución ¿cómo podemos afirmar que la solución sea unica?

Estas preguntas son fundamentales en matemáticas aplicadas, ciencias de la computación, ingeniería y física, pues de la respuesta depende si un problema está bien planteado o si puede ser resuelto mediante métodos algebraicos.


Para ahondar mas en el estudio de estas preguntas tomemos en cuenta los siguientes casos.


## **1. Solución única**
**Problema:**  
En una papelería, un cuaderno y dos plumas cuestan \$8. Dos cuadernos y una pluma cuestan \$7. ¿Cuánto cuesta cada artículo?

$$
\begin{cases}
x + 2y = 8 \\
2x + y = 7
\end{cases}
$$

donde $x$ es el precio del cuaderno y $y$ el de la pluma.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 8, 100)
y1 = (8 - x)/2       # x + 2y = 8  =>  y = (8 - x)/2
y2 = 7 - 2*x         # 2x + y = 7  =>  y = 7 - 2x

plt.figure(figsize=(6,6))
plt.plot(x, y1, label='$x + 2y = 8$')
plt.plot(x, y2, label='$2x + y = 7$')
plt.axhline(0, color='black', linewidth=2)  # Eje X más marcado
plt.axvline(0, color='black', linewidth=2)  # Eje Y más marcado
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Sistema con una solución única')
plt.grid(True)
plt.xlim(-1,8)
plt.ylim(-4,8)

# Encontramos la intersección
A = np.array([[1,2],[2,1]])
b = np.array([8,7])
sol = np.linalg.solve(A,b)
plt.scatter(sol[0], sol[1], color='red', zorder=5, label=f'Solución: ({sol[0]:.1f}, {sol[1]:.1f})')
plt.legend()
plt.show()
```

## **2. Infinitas soluciones**
**Problema:**  
En una frutería, el precio de una bolsa de manzanas más una bolsa de peras es \$4. Si un amigo compra el doble de bolsas de cada fruta, paga \$8. ¿Cuál es el precio de cada bolsa?

$$
\begin{cases}
x + y = 4 \\
2x + 2y = 8
\end{cases}
$$

donde $x$ es el precio de la bolsa de manzanas y $y$ el de la bolsa de peras.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 6, 100)
y1 = 4 - x           # x + y = 4
# Segunda ecuación es equivalente: 2x + 2y = 8 => x + y = 4 => y = 4 - x
y2 = 4 - x

plt.figure(figsize=(6,6))
plt.plot(x, y1, label='$x + y = 4$')
plt.plot(x, y2, '--', label='$2x + 2y = 8$ (misma recta)')
plt.axhline(0, color='black', linewidth=2)  # Eje X más marcado
plt.axvline(0, color='black', linewidth=2)  # Eje Y más marcado
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Sistema con infinitas soluciones')
plt.grid(True)
plt.xlim(-2,6)
plt.ylim(-2,6)
plt.legend()
plt.show()
```

## **3. Sin solución**
**Problema:**  
En una panadería, el cartel dice que una pieza de pan más una concha cuestan \$3. Al día siguiente, por error, el mismo cartel dice que la misma suma cuesta \$5. ¿Cuánto cuesta cada cosa?

$$
\begin{cases}
x + y = 3 \\
x + y = 5
\end{cases}
$$

donde $x$ es el precio de la pieza de pan y $y$ el de la concha.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 8, 100)
y1 = 3 - x           # x + y = 3
# x + y = 5 => y = 5 - x

plt.figure(figsize=(6,6))
plt.plot(x, y1, label='$x + y = 3$')
plt.plot(x, 5-x, label='$x + y = 5$')
plt.axhline(0, color='black', linewidth=2)  # Eje X más marcado
plt.axvline(0, color='black', linewidth=2)  # Eje Y más marcado
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Sistema sin solución')
plt.grid(True)
plt.xlim(-2,8)
plt.ylim(-2,8)
plt.legend()
plt.show()
```
