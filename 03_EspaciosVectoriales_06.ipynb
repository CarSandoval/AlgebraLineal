# 3.6 Propiedades de las matrices

## 3.6.1 Espacio nulo

### Definición

El **espacio nulo** (o **kernel**) de una matriz $A$ de tamaño $m \times n$ es el conjunto de todos los vectores $x \in \mathbb{R}^n$ (o en el campo correspondiente) tales que $A x = 0$:
$$
\text{Nul}(A) = \{ x \in \mathbb{R}^n \mid A x = 0 \}
$$
Es un subespacio vectorial de $\mathbb{R}^n$.



### Ejemplos

- Sea $A = \begin{pmatrix}1 & 2 \\ 2 & 4\end{pmatrix}$.  
  Buscamos $x = (x_1, x_2)$ tal que $A x = 0$:

  $$
  \begin{cases}
    x_1 + 2x_2 = 0 \\
    2x_1 + 4x_2 = 0
  \end{cases}
  \implies x_1 = -2x_2
  $$
  El espacio nulo es $\{ t(-2,1) \mid t \in \mathbb{R} \}$.



### Propiedades

- El espacio nulo siempre es un subespacio de $\mathbb{R}^n$.
- $A x = 0$ siempre tiene al menos la solución trivial ($x = 0$).
- El conjunto de soluciones del sistema homogéneo $A x = 0$ es el espacio nulo.
- Su dimensión se llama **nulidad** de la matriz.


## 3.6.2. Nulidad

### Definición

La **nulidad** de una matriz $A$ es la **dimensión** de su espacio nulo:
$$
\text{nulidad}(A) = \dim(\text{Nul}(A))
$$



### Ejemplos

- En el ejemplo anterior, el espacio nulo es una recta en $\mathbb{R}^2$, por lo que la nulidad es 1.
- Si $A$ es la identidad, su espacio nulo es $\{0\}$, así que la nulidad es 0.



### Propiedades

- $0 \leq \text{nulidad}(A) \leq n$, donde $n$ es el número de columnas de $A$.
- La suma de la nulidad y el rango de $A$ es igual al número de columnas:
  $$
  \text{nulidad}(A) + \text{rango}(A) = n
  $$



## 3.6.3. Imagen

### Definición

La **imagen** de una matriz $A$ es el conjunto de todos los vectores $y \in \mathbb{R}^m$ para los cuales existe $x \in \mathbb{R}^n$ tal que $A x = y$:
$$
\text{Im}(A) = \{ y \in \mathbb{R}^m \mid \exists x \text{ tal que } A x = y \}
$$
También se llama **espacio columna** de $A$.



### Ejemplos

Si $A = \begin{pmatrix}1 & 2 \\ 2 & 4\end{pmatrix}$, todas las columnas son múltiplos de $(1,2)$.  
  La imagen es $\{ t(1,2) \mid t \in \mathbb{R} \}$.



### Propiedades

- La imagen es un subespacio de $\mathbb{R}^m$.
- Su dimensión es igual al rango de la matriz.
- Todos los posibles resultados de $A x$ pertenecen a la imagen.



## 3.6.4. Rango

### Definición

El **rango** de una matriz $A$ es el número máximo de columnas linealmente independientes (o filas linealmente independientes).  
Equivale a la dimensión del espacio columna o del espacio renglón de $A$.

$$
\text{rango}(A) = \dim(\text{Im}(A))
$$



### Ejemplos

- Para $A = \begin{pmatrix}1 & 2 \\ 2 & 4\end{pmatrix}$, solo una columna es linealmente independiente: el rango es 1.
- Para $A = \begin{pmatrix}1 & 0 \\ 0 & 1\end{pmatrix}$, las dos columnas son independientes: el rango es 2.



### Propiedades

- $0 \leq \text{rango}(A) \leq \min\{m, n\}$
- El rango de $A$ es igual al número de pivotes en la reducción escalonada.
- El rango es igual tanto para columnas como para filas:  
  $\text{rango}(A) = \text{rango}(A^T)$.


## 3.6.5. Espacio de las columnas

### Definición

El **espacio de las columnas** de $A$ es el subespacio de $\mathbb{R}^m$ generado por las columnas de $A$.  
Es igual a la imagen de $A$.



### Ejemplos

Considera la matriz:
$$
A = \begin{pmatrix}
1 & 2 & 3 \\
0 & 1 & 2 \\
2 & 4 & 6 \\
\end{pmatrix}
$$

Las columnas de $A$ son:
- $C_1 = (1, 0, 2)$
- $C_2 = (2, 1, 4)$
- $C_3 = (3, 2, 6)$



**¿Son todas las columnas linealmente independientes?**

Supón que $C_3 = aC_1 + bC_2$:

Planteamos el sistema:
- Primera componente: $3 = a \cdot 1 + b \cdot 2$
- Segunda componente: $2 = a \cdot 0 + b \cdot 1$
- Tercera componente: $6 = a \cdot 2 + b \cdot 4$

De la segunda ecuación: $2 = b \implies b = 2$

Sustituimos en la primera:
$3 = a + 2 \cdot 2 \implies 3 = a + 4 \implies a = -1$

Verificamos la tercera:
$6 = 2a + 4b = 2(-1) + 4 \cdot 2 = -2 + 8 = 6$



**Conclusión:**
- $C_3 = -1 \cdot C_1 + 2 \cdot C_2$
- Por lo tanto, **$C_1$ y $C_2$ son linealmente independientes**, y $C_3$ es combinación lineal de ellos.

El **espacio de las columnas** de $A$ es:
$$
\text{Col}(A) = \text{span}\{C_1, C_2\}
$$
Es un subespacio de dimensión 2 en $\mathbb{R}^3$ (es un plano que pasa por el origen).



**¿Cómo se ve ese espacio?**  
Cualquier vector columna de $A$ es de la forma $aC_1 + bC_2$ para algunos $a, b \in \mathbb{R}$:
$$
a(1, 0, 2) + b(2, 1, 4) = (a + 2b,\; b,\; 2a + 4b)
$$

Es decir, el espacio de las columnas es el conjunto de todos los vectores $(x, y, z)$ tales que existen $a, b$ con:
- $x = a + 2b$
- $y = b$
- $z = 2a + 4b$

Podemos eliminar $a$:
- $a = x - 2b$
- $z = 2(x - 2b) + 4b = 2x - 4b + 4b = 2x$

Así, **el espacio de las columnas es el conjunto de todos los vectores $(x, y, z)$ en $\mathbb{R}^3$ tales que $z = 2x$**.



**Resumen:**  
El espacio de las columnas de $A$ es un plano en $\mathbb{R}^3$ dado por $z = 2x$,  
y su **dimensión es 2**.


### Propiedades

- Es un subespacio de $\mathbb{R}^m$.
- Su dimensión es el rango de $A$.
- Cualquier combinación lineal de las columnas de $A$ pertenece a este espacio.

## 3.6.6. Espacio de los renglones


### Definición

El **espacio de los renglones** de una matriz $A$ es el subespacio de $\mathbb{R}^n$ generado por las filas de $A$ (consideradas como vectores fila).



### Ejemplos

Considera la matriz
$$
A = \begin{pmatrix}
1 & 2 & 3 \\
2 & 4 & 6 \\
0 & 1 & 1 \\
\end{pmatrix}
$$

Sus **renglones** son:
- $R_1 = (1, 2, 3)$
- $R_2 = (2, 4, 6)$
- $R_3 = (0, 1, 1)$



**¿Son los renglones linealmente independientes?**

Observa que $R_2 = 2R_1$.  
Por lo tanto, $R_1$ y $R_2$ son linealmente dependientes.

Ahora verifica $R_1$ y $R_3$:

Supón que existe $\alpha$ tal que $R_3 = \alpha R_1$:

- Primera componente: $0 = \alpha \cdot 1 \implies \alpha = 0$
- Segunda componente: $1 = \alpha \cdot 2 \implies 1 = 0$ (no es posible)

Así que **$R_3$ no es múltiplo de $R_1$**.  
Por lo tanto, $R_1$ y $R_3$ son linealmente independientes.



**Espacio de los renglones:**

El espacio de los renglones es el subespacio generado por $R_1$ y $R_3$:
$$
\text{Row}(A) = \text{span}\{(1, 2, 3),\, (0, 1, 1)\}
$$
que es un plano en $\mathbb{R}^3$.



**Conclusión:**  
- El **espacio de los renglones** de $A$ es de dimensión $2$ en $\mathbb{R}^3$.
- Una base es $\{(1, 2, 3),\, (0, 1, 1)\}$.



### Propiedades

- Es un subespacio de $\mathbb{R}^n$.
- Su dimensión es el rango de $A$.
- El espacio de los renglones de $A$ es igual al espacio de las columnas de $A^T$ (la traspuesta).
