# 2.1 El conjunto de permutaciones  $ S_n $ 

El conjunto  $ S_n $  (llamado el **grupo simétrico** de grado  $ n $ ) es el conjunto de todas las **permutaciones** posibles de  $ n $  objetos. Es decir, es el conjunto de todas las formas de reordenar una secuencia de  $ n $  elementos. Donde Cada elemento de  $ S_n $  se llama **permutación**.

## 2.1.1 ¿Cómo se ven los elementos en  $ S_n $ ?

Una permutación se puede describir de varias maneras:

- **Notación de dos filas:**  
  Una permutación  $ \sigma $  en  $ S_n $  se representa como:

  $$
  \sigma = 
  \begin{pmatrix}
  1 & 2 & 3 & \cdots & n \\
  \sigma(1) & \sigma(2) & \sigma(3) & \cdots & \sigma(n)
  \end{pmatrix}
  $$

  Por ejemplo, en  $ S_3 $ , la permutación que envía  $ 1 \mapsto 2 $ ,  $ 2 \mapsto 1 $ ,  $ 3 \mapsto 3 $  es:

  $$
  \begin{pmatrix}
  1 & 2 & 3 \\
  2 & 1 & 3
  \end{pmatrix}
  $$

- **Notación ciclo:**  
  Alternativamente, las permutaciones pueden escribirse como productos de ciclos, por ejemplo:  
   $ (1\ 2\ 3) $  significa  $ 1 \mapsto 2 $ ,  $ 2 \mapsto 3 $ ,  $ 3 \mapsto 1 $ .


## 2.1.2 Ejemplo:  $ S_3 $ 

 $ S_3 $  tiene  $ 3! = 6 $  elementos:

1. Identidad:  $ \begin{pmatrix}1&2&3\\1&2&3\end{pmatrix} $ 
2.  $ (1\ 2) $ :  $ \begin{pmatrix}1&2&3\\2&1&3\end{pmatrix} $ 
3.  $ (1\ 3) $ :  $ \begin{pmatrix}1&2&3\\3&2&1\end{pmatrix} $ 
4.  $ (2\ 3) $ :  $ \begin{pmatrix}1&2&3\\1&3&2\end{pmatrix} $ 
5.  $ (1\ 2\ 3) $ :  $ \begin{pmatrix}1&2&3\\2&3&1\end{pmatrix} $ 
6.  $ (1\ 3\ 2) $ :  $ \begin{pmatrix}1&2&3\\3&1&2\end{pmatrix} $ 

## 2.1.3 Composición de dos permutaciones

La **composición** de dos permutaciones  $ \sigma $  y  $ \tau $  en  $ S_n $ , denotada  $ \sigma \circ \tau $ , consiste en aplicar primero  $ \tau $  y después  $ \sigma $ :

$$
(\sigma \circ \tau)(i) = \sigma(\tau(i))
$$

> Es importante notar que la composición de permutaciones no es conmutativa en general:  $ \sigma \circ \tau \neq \tau \circ \sigma $ .

**Ejemplo:**  
Si:

-  $ \sigma = \begin{pmatrix}1&2&3\\2&1&3\end{pmatrix} $ 
-  $ \tau = \begin{pmatrix}1&2&3\\1&3&2\end{pmatrix} $ 

Entonces,
-  $ (\sigma \circ \tau)(1) = \sigma(\tau(1)) = \sigma(1) = 2 $ 
-  $ (\sigma \circ \tau)(2) = \sigma(\tau(2)) = \sigma(3) = 3 $ 
-  $ (\sigma \circ \tau)(3) = \sigma(\tau(3)) = \sigma(2) = 1 $ 

Así, 
$$
\sigma \circ \tau = \begin{pmatrix}1&2&3\\2&3&1\end{pmatrix}
$$

## 2.1.4 Propiedades de  $ S_n $ 

-  $ S_n $  tiene  $ n! $  (factorial de  $ n $ ) elementos.
- La composición de dos permutaciones es otra permutación en  $ S_n $ .
- Existe un elemento neutro (la permutación identidad, que no cambia nada).
- Cada permutación tiene un inverso, que deshace el reordenamiento.
-  $ S_n $  con la operación de composición es un **grupo** (es decir, cumple las propiedades de grupo: asociatividad, elemento neutro, existencia de inverso).
- Para  $ n \geq 3 $ ,  $ S_n $  no es abeliano (el orden de la composición sí importa).

## 2.1.5 Definición de Signo de una permutación ($\mathrm{sgn}(\sigma)$)

El **signo** de una permutación $\sigma$ nos dice si la permutación es **par** ($+1$) o **impar** ($-1$). Decimos que: 

$$
\mathrm{sgn}(\sigma) = (-1)^{\text{número de inversiones de }\ \sigma}
$$,  
donde una **inversión** es un par $(i,j)$ con $i<j$ pero $\sigma(i) > \sigma(j)$.

**Ejemplo:**  
La permutación $(2,1,3)$ en $S_3$ tiene una inversión: $2>1$ en la posición $(1,2)$,  
por lo tanto es impar y $\mathrm{sgn}((2,1,3)) = -1$.

## 2.1.6 Definición de transposición

Una **transposición** es una permutación que intercambia exactamente dos elementos y deja todos los demás fijos.

En  $ S_n $  (el conjunto de todas las permutaciones de  $ n $  elementos), la transposición que intercambia las posiciones  $ i $  y  $ j $  (con  $ i \neq j $ ) se denota como  $ (i\ j) $ .

**Ejemplo:**

En  $ S_4 $ , la transposición  $ (2\ 4) $  intercambia los elementos en las posiciones 2 y 4:

$$
\begin{pmatrix}
1 & 2 & 3 & 4 \\
1 & 4 & 3 & 2
\end{pmatrix}
$$

- El 2 pasa a la posición 4.
- El 4 pasa a la posición 2.
- Los elementos 1 y 3 permanecen en sus posiciones.

## 2.1.7 Demostración de que la función de signo esta bien definida

Probar que el signo está bien definido significa demostrar que la paridad (par o impar) del número de transposiciones necesarias para formar una permutación es siempre la misma, sin importar la secuencia de transposiciones que elijas.

Usando la definición de signo basada en **inversiones**, podemos demostrarlo de la siguiente manera. 



**Lema:**

Si $\sigma$ es una permutación y $\tau$ es una transposición, entonces el número de inversiones de $\tau \circ \sigma$ y el número de inversiones de $\sigma$ **tienen paridad diferente**. Es decir:

$$
\text{inv}(\tau \circ \sigma) = \text{inv}(\sigma) + \text{un número impar}
$$

**Demostración del Lema:**

1.  Primero, consideremos una **transposición simple** que intercambia elementos adyacentes, $\tau = (i, i+1)$. Cuando componemos $\sigma$ con $\tau$, lo único que cambia es el orden de los valores que $\sigma$ tenía en las posiciones $i$ e $i+1$. El único par cuya relación de orden se invierte es precisamente $(\sigma(i), \sigma(i+1))$. Por lo tanto, el número de inversiones cambia exactamente en $+1$ o $-1$. En ambos casos, la paridad cambia.

2.  Ahora, cualquier transposición $\tau = (a, b)$ con $a < b$ puede escribirse como un producto de un **número impar** de transposiciones adyacentes. Para llevar el elemento en la posición $a$ hasta la posición $b$ se necesitan $(b-a)$ intercambios adyacentes. Luego, para llevar el elemento que estaba en $b$ de vuelta a la posición $a$ se necesitan $(b-a-1)$ intercambios. El total es:

$$ 
(b-a) + (b-a-1) = 2(b-a) - 1 
$$ 

Notemos que este número es siempre **impar**.

3.  Como multiplicar por $(a,b)$ es lo mismo que multiplicar un número impar de transposiciones adyacentes, y cada una de ellas cambia la paridad de las inversiones, el efecto neto es un cambio de paridad. El número total de inversiones cambia por una suma de un número impar de $\pm 1$, lo que siempre resulta en un número impar.

**Demostración Principal**

Ahora, usemos el lema anterior para la prueba final.

1.  Supongamos que una permutación $\sigma$ puede escribirse como un producto de $k$ transposiciones:
    $$\sigma = \tau_k \circ \dots \circ \tau_2 \circ \tau_1$$

2.  Podemos ver esto como si partiéramos de la permutación identidad ($id$), que tiene **0 inversiones** (0 es par), y le aplicáramos una a una las transposiciones.

3.  Veamos cómo cambia el número de inversiones:
    * $\text{inv}(id) = 0$ (par)
    * $\text{inv}(\tau_1 \circ id)$ = $0 + \text{impar}$ = **impar**
    * $\text{inv}(\tau_2 \circ \tau_1 \circ id)$ = $\text{impar} + \text{impar}$ = **par**
    * $\text{inv}(\tau_3 \circ \tau_2 \circ \tau_1 \circ id)$ = $\text{par} + \text{impar}$ = **impar**

4.  Por inducción, después de aplicar $k$ transposiciones, la paridad del número de inversiones de $\sigma$ será **la misma que la paridad de $k$**.
    $$\text{inv}(\sigma) \equiv k \pmod{2}$$

5.  Ahora, si $\sigma$ también tuviera otra descomposición con $m$ transposiciones, el mismo argumento nos diría que:
    $$\text{inv}(\sigma) \equiv m \pmod{2}$$

6.  Por lo tanto, tenemos que:
    $$k \equiv \text{inv}(\sigma) \equiv m \pmod{2} \implies k \equiv m \pmod{2}$$
    Esto demuestra que $k$ y $m$ **deben tener la misma paridad**.


De lo anterio concluimos que cualquier descomposición de una permutación $\sigma$ en transposiciones siempre tendrá un número de elementos con la misma paridad, el signo definido como $(-1)^{\text{número de transposiciones}}$ dará siempre el mismo resultado.

Por lo tanto, la función signo está **bien definida**.

## 2.1.8 Algunas propiedades de la función  $ \mathrm{sgn}(\sigma) $  - signo de una permutación

- **Valores posibles:**  
  Para cualquier permutación  $ \sigma \in S_n $ ,

  $$
  \mathrm{sgn}(\sigma) = \begin{cases}
  +1, & \text{si  $ \sigma $  es par} \\
  -1, & \text{si  $ \sigma $  es impar}
  \end{cases}
  $$

- **Producto de permutaciones:**  
  El signo de la composición de dos permutaciones es el producto de sus signos:
  
  $$
  \mathrm{sgn}(\sigma \circ \tau) = \mathrm{sgn}(\sigma) \cdot \mathrm{sgn}(\tau)
  $$

- **Transposición:**  
  El signo de una transposición es siempre  $ -1 $ .

- **Identidad:**  
  El signo de la permutación identidad es  $ +1 $ .

- **Inverso:**  
  El signo de la inversa de una permutación es igual al signo de la permutación:
  
  $$
  \mathrm{sgn}(\sigma^{-1}) = \mathrm{sgn}(\sigma)
  $$
